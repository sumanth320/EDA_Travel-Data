```python
import pandas as pd
import seaborn as sns
import datetime as dt
pd.options.display.float_format = '{:.5f}'.format
```

```python
df = pd.read_csv(r'C:\Users\Sumanth\Desktop\q1.csv')
df.head()
```

```python
#Checking the null values in our dataset
sns.heatmap(df.isnull(),yticklabels=False, cbar = False)
```

```python
df.isnull().mean()*100
#the null values are only 4.8% which is a very minor percentage hence these rows can be dropped
```

```python
df.dropna(inplace=True)
```

```python
#rounding off large decimal numbers to 2 places to easily read and manipulate data
df['net_revenue']=df['net_revenue'].round(2)
df['gross_merchandise_value']=df['gross_merchandise_value'].round(2)
```

```python
df
```

```python
df['percentage']=df['net_revenue']/df['gross_merchandise_value']
```

```python
#grouping data by Activity Category and then adding the GMV and NR values
df_new = df.groupby(['activity_category']).sum()[["net_revenue", "gross_merchandise_value"]]
```

```python
df_new
```

```python
#calculating NR as a percentage of GMV to understand our running costs per category
df_new['percentage']=(df_new['net_revenue']/df_new['gross_merchandise_value'])*100
```

```python
df_new.sort_values(by="gross_merchandise_value",ascending=False)
```

```python
df_new.sort_values(by="percentage",ascending=False)
```
