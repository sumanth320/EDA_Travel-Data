```python
import pandas as pd
import seaborn as sns
import datetime as dt
pd.options.display.float_format = '{:.5f}'.format
```

```python
df = pd.read_csv(r'C:\Users\Sumanth\Desktop\q2.csv')
df
```

```python
#cleaning data by rounding off numbers to 2 decimal places
df= df.round({'gmv_a': 2, 'gmv_after_coupons_a': 2,'commission_rate_a':2,'exchange_rate_a':2,'gmv_after_coupons_a':2})
```

```python
df
```

```python
#checking null value percentage
df.isnull().mean()*100
#coupon_information column can be dropped off since close to 60% values are null
#other rows with null values can be dropped off since only 0.44% are null.
```

```python
df=df.drop(['coupon_information'], axis = 1)
df
#dropping this column since this column has 58% of null values
```

```python
df.dropna(inplace=True)
```

```python
#converting columns to required date time format
pd.to_datetime(df["date_of_creation"])
df['date_of_checkout'] = pd.to_datetime(df.date_of_checkout, format='%Y-%m-%d %H:%M:%S')
df['date_of_creation'] = pd.to_datetime(df.date_of_creation, format='%Y-%m-%d %H:%M:%S')
```

```python
#extracting seconds from the columns to caluclate the average time taken to checkout
df['totsec'] = (df['date_of_checkout']-df["date_of_creation"]).astype('timedelta64[s]')
df=df[df['totsec']>0]
#values less than 0 have been removed since this is clearly an error since end date cannot be before start date for shopping cart
```

```python
print(df['totsec'].describe())
(df['totsec'].sum()/len(df['totsec']))/60
# calculating the average time for checkout. due to outliers this value may increase.
#29.06 mins is the average time for shopping
```

```python
dffr = df[df['is_fraud']==True].groupby(['coupon_segment_id_a']).size().reset_index()
dffr['percentage of total bookings'] = (dffr[0]/len(df['is_fraud']))*100
dffr.sort_values(by=0,ascending=False)
#most number of fraudulent cases are from a cetain coupon segment which requires further investigation into that coupon
```

```python
dffr1 = df[df['is_fraud']==True].groupby(['reseller_id_a']).size().reset_index()
dffr1['percentage of total bookings'] = (dffr1[0]/len(df['is_fraud']))*100
dffr1.sort_values(by=0,ascending=False)
#most number of fraudulent cases are from 2 Partners. These are suspicous activities and these partners need to be looked into
```

```python
dffr2 = df[df['is_fraud']==True].groupby(['gyg_campaign_id_a']).size().reset_index()
dffr2['percentage of total bookings'] = (dffr2[0]/len(df['is_fraud']))*100
dffr2.sort_values(by=0,ascending=False)
#most number of fraudulent cases are from 1 specific GYG campaign. Needs to be investigated
```

```python
dffr3 = df.groupby(['gyg_campaign_id_a']).size().reset_index().sort_values(by=0,ascending=False)
dffr3['percentage of total bookings'] = (dffr3[0]/len(df['gyg_campaign_id_a']))*100
dffr3.head()
#most number of bookings are from the GYG campaign -1670924195 and this is the same campaign which has the highest frauds.
#High fraud count might be because of the large number of bookings and might not be related to that campaign. 
#But it needs to be investigated
```
